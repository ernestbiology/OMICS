# Проект 1 — Анализ Hi-C

</div>

---

## Цель

Цель проекта — обработать данные Hi-C и сравнить контактные карты между четырьмя образцами: **MoPh7**, **MoPh11**, **MoPh14** и **MoPh15**.

`MoPh7` использовался как референсный/контрольный образец для визуального сравнения.

---

## Краткое описание воркфлоу

Основной воркфлоу включал:

1. Обработку парных ридов Hi-C (paired-end).
2. Построение `.hic`-файлов.
3. Открытие контактных карт в **Juicebox**.
4. Сравнение паттернов контактов на уровне полного геном и отдельных хромосом.

---

## Результат

> Полногеномные карты Hi-C показали ожидаемый сильный диагональный сигнал во всех образцах. Чётких крупных вне-диагональных межхромосомных блоков не обнаружено.

---

## Биологическая интерпретация

Результат говорит о том, что на этих картах Hi-C не видно явных крупномасштабных межхромосомных перестроек относительно `MoPh7`.

---

## Возможные проблемы

Глубина покрытия в полученных картах недостаточна для надёжного анализа тонких структур — **TADs**, **петель (loops)** и небольших перестроек. Для их выявления потребовалось бы более глубокое секвенирование и дополнительный статистический анализ; на текущих данных такие выводы делать некорректно.
