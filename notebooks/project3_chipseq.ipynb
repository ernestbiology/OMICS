# Проект 3 — Анализ ChIP-seq

---

## Цель

Цель проекта — обработать данные ChIP-seq для хроматиновых меток/факторов и выявить пики обогащения относительно input.

Обработанные образцы:

* MoPh7_H3K9me3
* MoPh7_H3K27Ac
* MoPh7_H2A119Ub
* MoPh7_CTCF
* input control

---

## Краткое описание воркфлоу

Воркфлоу включал:

1. Скачивание исходных ридов ChIP-seq.
2. Обрезку адаптеров (adapter trimming).
3. Выравнивание BWA.
4. Фильтрацию по MAPQ.
5. Генерацию bigWig-треков.
6. Вызов пиков с помощью MACS3.
7. Генерацию треков fold-enrichment.
8. Визуализацию в IGV.

---

## Выходные файлы

Пайплайн сгенерировал:

* q30 bigWig-треки
* input bigWig-трек
* файлы пиков MACS3
* bigWig-треки fold-enrichment
* summit BED-файлы для narrow-пиков

---

## Интерпретация

Сигнал ChIP-seq для CTCF сильно обогащён относительно input. MACS3 выявляет в этом регионе узкий пик, а summit отмечает точку максимального обогащения.

Это согласуется с локальным сайтом связывания CTCF.
